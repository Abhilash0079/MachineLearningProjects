# <b>COVID-19 Forecasting & Analytics Platform</b>

## <b>Model Analysis</b>

### <b>Objective</b>

The objective of this notebook is to evaluate and compare machine learning models for forecasting daily COVID-19 cases in India.

Two forecasting models were developed:

1. Random Forest Regressor
2. XGBoost Regressor

The best-performing model will be selected for deployment in the dashboard application.

### <b>Random Forest Results</b>

#### <b>Model Performance</b>

| Metric | Value |
|----------|----------|
| MAE | 2358.96 |
| RMSE | 20522.59 |
| R² | 0.9837 |

#### <b>Interpretation</b>

The Random Forest model achieved strong predictive performance, explaining approximately 98.37% of the variance in daily COVID-19 cases.

### <b>Random Forest Feature Importance</b>

| Feature | Importance |
|----------|----------|
| lag_7 | 86.38% |
| cases_7d_avg | 8.99% |
| cases_14d_avg | 2.97% |
| mortality_rate | 0.53% |
| lag_14 | 0.44% |
| vaccination_rate | 0.37% |
| month | 0.31% |
| weekday | ~0% |
| lag_1 | ~0% |

#### <b>Key Finding</b>

The model relies heavily on the 7-day lag feature, indicating strong weekly seasonality in COVID-19 case reporting.

### <b>XGBoost Results</b>

#### <b>Model Performance</b>

| Metric | Value |

#### <b>Interpretation</b>

XGBoost achieved the highest explanatory power among the evaluated models and demonstrated superior forecasting performance.

### <b>Model Comparison</b>

| Metric | Random Forest | XGBoost |

### <b>Analysis</b>

#### <b>Random Forest</b>

### <b>Conclusion</b>

XGBoost was selected as the final production model due to its superior predictive performance.

### <b>Final Model Selection</b>

Selected Model: XGBoost Regressor

#### <b>Reasons</b>

- Highest R² Score
- Lowest RMSE
- Better handling of nonlinear relationships
- Suitable for deployment in the forecasting dashboard

The XGBoost model will be used as the forecasting engine in the final COVID-19 Analytics & Forecasting Dashboard.
